## This File contains -:
1) Python code for Conversion to valid json lines.<br />
2) code for Cleaning of double quotes inside "Text" field. <br />
3) Code to upload jsonl file on openAI.<br />

```python
import pandas as pd
import numpy as np

with open("C:\\Users\\shrey\\OneDrive\\Documents\\RMIT\\Semester 4\\Data Science postgraduate project\\Lexata\\RMIT-Summer-2021-main\\all_data_for_openai.json",errors="ignore") as JsonData:
    
    with open("C:\\Users\\shrey\\OneDrive\\Documents\\RMIT\\Semester 4\\Data Science postgraduate project\\Lexata\\RMIT-Summer-2021-main\\Modified__data_for_openai.jsonl",'w',errors="ignore") as Outputfile:
        
        for row in JsonData:
            row=row.replace('â€œ','') # Open double Quotation mark
            row=row.replace('â„¢','') # apostrophe mark
            row=row.replace('â€','') # close double quotation mark
            row=row.replace('"},','"}')
            row=row.replace('">',' ')
            row=row.replace("\n","")
            Outputfile.write(row)
            Outputfile.write("\n")
```
